# Table 3. Distribution of marker tags in PRs generated by Copilot for PRs.

```python
import re
summary_command = 0
all_command = 0
walkthrough_command = 0
poem_command = 0
commands = []
for index, row in valid_df.iterrows():
    command = list(set(re.findall(r"\bcopilot:(all|summary|walkthrough|poem)\b", row['body'])))
    if 'summary' in command: summary_command += 1
    if 'all' in command: all_command += 1
    if 'walkthrough' in command: walkthrough_command += 1
    if 'poem' in command: poem_command += 1
    if len(command) == 0:
        print(row['url'], row['body'])
    commands.append(command)
```

```python
print(summary_command,
all_command,
walkthrough_command,
poem_command)
```

```python
count = Counter(tuple(item) for item in commands)
sumz = 0
# Display the result
for combination, freq in count.items():
    sumz += freq
    print(f"Combination: {combination}, Frequency: {freq}, sum : {sumz}")
```
